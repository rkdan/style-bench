## Plotting a single analysis
Here, we plot the analyses of press conference question and answer pairs from former President George W. Bush, generated from the `sample-analyze.yaml` config file.

````python
import pickle

def load_pickle(file_path):
    with open(file_path, 'rb') as file:
        return pickle.load(file)

bush = load_pickle('output/bush/bush-100_2025-08-18_09-51-32/results.pkl')
obama = load_pickle('output/obama/obama-100_2025-08-18_10-46-20/results.pkl')
````

```yaml
function_word_frequency: list

    sadness : list
    surprise : list
```
